```python
from transformers import AutoTokenizer, PhiForCausalLM
from google.colab import files
import torch
import ast
uploaded = files.upload()
text_file_path = list(uploaded.keys())[0]

with open(text_file_path, "r", encoding="latin-1") as file:
    file_contents = file.read()

# Convert the file content to a Python list
students_list = ast.literal_eval(file_contents)
device = "cuda" if torch.cuda.is_available() else "cpu"
model = PhiForCausalLM.from_pretrained("microsoft/phi-1").to(device)
tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-1")


def process_students_in_batches(students, batch_size=10):
    for i in range(0, len(students), batch_size):
        # Select a batch of 25 students
        batch = students[i:i + batch_size]

        # Define a clear and direct prompt for the model
        prompt = (
            "The following is a list of student performance data:\n"
            f"{batch}\n"
            "Please analyze this data and identify patterns and factors contributing to student success. "
            "For each student, provide personalized learning recommendations with a focus on the following aspects:\n"
            "1. Areas where the student is underperforming and needs improvement.\n"
            "2. Specific subjects where the student is excelling.\n"
            "3. Actionable steps the student can take to improve their performance in underperforming areas.\n"
            "4. A motivational message for the student to encourage better performance."
        )

        # Generate analysis with your model
        inputs = tokenizer(prompt, return_tensors="pt").to(device)
        generate_ids = model.generate(inputs.input_ids, max_length=1000)
        tokenizer.batch_decode(generate_ids, skip_special_tokens=True, clean_up_tokenization_spaces=False)[0]
        generated_text=tokenizer.batch_decode(generate_ids, skip_special_tokens=True, clean_up_tokenization_spaces=False)[0]



        with open("analysis_using_phi-1.txt", "a", encoding="utf-8") as file:
            file.write(f"{generated_text} \n\n")
        # Write the result to a file using utf-8 encoding


# Process the entire student list in chunks of 25
process_students_in_batches(students_list)
```
